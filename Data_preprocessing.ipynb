MISSING VALUE TREATMENTS


```python
for i in ["mileage(kmpl)","engine(cc)","torque(Nm)","max_power(bhp)"]:
  data[i].fillna(data[i].median(),inplace=True)
```

```python
#choose the method of imputing missing values
#like mean, median, mode
data.isnull().sum()
```
